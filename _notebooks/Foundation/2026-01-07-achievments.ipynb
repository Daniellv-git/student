Sprint 1
In my first sprint, I got familiar with the development environment by setting up a repository using VS Code and GitHub, and getting my local host site running. Once I had everything connected and syncing properly, I started building out my portfolio with things like an About Me page and learning how to manage issues on GitHub. From there I jumped into creating games, including versions with connect 4, which was a great way to get hands on experience while also picking up basic programing.
Sprint 2
Sprint 2 was all about diving deep into the core concepts of JavaScript and programming as a whole. I worked through topics like loops and iterations, conditionals, arrays, booleans, JSON, data abstraction, strings, mathematical expressions, classes and methods, and JavaScript with the HTML DOM. Each topic built on the last and helped me see how everything connects. I also got experience working with Jupyter Notebooks, learning about things like permalinks, front matter, markdown and JavaScript cells.
Sprint 3
In Sprint 3, my team and I built an RPG game using a Game Runner and GameEngine framework. My personal contributions included designing background and doing the coding of the game.
