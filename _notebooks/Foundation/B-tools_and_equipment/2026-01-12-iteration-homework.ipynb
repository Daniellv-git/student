 _Here is an example of one of the loops that was went over in class:_


```python
let fruits = ["Heart Shaped Herb", "Yami Yami no Mi", "Gomu Gomu no Mi"];

// Loop 1: Normal for loop (goes forward)
console.log("Loop 1:");
for (let i = 0; i < fruits.length; i++) {
    console.log(fruits[i]);
}

// Loop 2: Reverse for loop (goes backward)
console.log("\nLoop 2:");
for (let i = fruits.length - 1; i >= 0; i--) {
    console.log(fruits[i]);
}

// Loop 3: While loop
console.log("\nLoop 3:");
let j = 0;
while (j < fruits.length) {
    console.log(fruits[j]);
    j++;
}
```
