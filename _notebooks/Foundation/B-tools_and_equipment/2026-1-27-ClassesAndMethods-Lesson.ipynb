class Cat {
    constructor(name) {
        this.name = name;
    }

    eat() {
        console.log(this.name + " is eating food!");
    }

    meow() {
        console.log(this.name + " says: Meow!");
    }
}

let myCat = new Cat("Whiskers");
myCat.eat();
myCat.meow();
