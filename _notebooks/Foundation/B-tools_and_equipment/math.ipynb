```python
// Homework Problem #4: Check if y is divisible by 5

let y = 20; // You can change this value to test different numbers

let remainder = y % 5;

console.log(y + " % 5 = " + remainder);

if (y % 5 === 0) {
    console.log(y + " is divisible by 5");
} else {
    console.log(y + " is NOT divisible by 5");
}
```
